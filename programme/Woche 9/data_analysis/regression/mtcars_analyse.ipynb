### Startpunkt

* Es besteht ein Zusammenhang zwischen Kraftstoffverbrauch und dem Gewicht des Autos

```python
import pandas as pd
```

```python
mtcars = pd.read_csv('mtcars.csv')
mtcars.head()
```

```python
mtcars.plot(kind='scatter', x='wt', y='mpg') 
```

### Vorhersagen auf Basis eines linearen Modells

* Imoport eines Science Kits für Machine Learning

```python
from sklearn import linear_model

```

```python
regression_model = linear_model.LinearRegression()
regression_model.fit(X = pd.DataFrame(mtcars['wt']), y=mtcars['mpg'])
print(regression_model.intercept_, regression_model.coef_)
```

```python
prediction = regression_model.predict(pd.DataFrame(mtcars['wt']))
```

```python
import matplotlib.pyplot as plt
mtcars.plot(kind='scatter', x='wt', y='mpg', color='green')
plt.plot(mtcars['wt'].values, prediction, color='red')
```

* Wie viel Anteil an der mpg-Änderung hat denn das Gewicht des Autos

```python
regression_model.score(X=pd.DataFrame(mtcars['wt']), y=mtcars['mpg'])
```

### Abhängigkeit zum Hubraum

```python
mtcars.plot(kind='scatter', x='disp', y='mpg', color='green')
```

* Analyse

```python
regression_model = linear_model.LinearRegression()
x_values = pd.DataFrame(mtcars['disp'])
regression_model.fit(X = x_values, y=mtcars['mpg'])
print(regression_model.intercept_, regression_model.coef_, regression_model.score(X=x_values, y=mtcars['mpg']))
prediction = regression_model.predict(x_values)
mtcars.plot(kind='scatter', x='disp', y='mpg', color='green')
plt.plot(mtcars['disp'].values, prediction, color='red')
```
